```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
names = []
ratings = []
reviews = []
ctype= []
hq = []
old = []
employees = []
#desc = []
final_df = pd.DataFrame()
for j in range(1,334):
    url = 'https://www.ambitionbox.com/list-of-companies?page={}'.format(j)
    webpage = requests.get(url).text
    soup = BeautifulSoup(webpage, 'lxml')
    company = soup.find_all('div',class_='company-content-wrapper')
    
    for i in company:
        try:
            names.append(i.find('h2').text.strip())
        except:
            names.append(np.nan)
        try:
            ratings.append(i.find('p', class_='rating').text.strip())
        except:
            ratings.append(np.nan)
        try:
            reviews.append(i.find('a',class_='review-count').text.strip())
        except:
            ratings.append(np.nan)
        try:
            ctype.append(i.find_all('p', class_='infoEntity')[0].text.strip())
        except:
            ctype.append(np.nan)
        try:
            hq.append(i.find_all('p', class_='infoEntity')[1].text.strip())
        except:
            hq.append(np.nan)
        try:
            old.append(i.find_all('p', class_='infoEntity')[2].text.strip())
        except:
            old.append(np.nan)
        try:
            employees.append(i.find_all('p', class_='infoEntity')[3].text.strip())
        except:
            employees.append(np.nan)
        #desc.append(i.find('p', class_='description').text.strip())
    
    d = {'name':names,'rating':ratings,'reviews':reviews,'type':ctype,'headquarter':hq,'old':old,'employees':employees}
    df_temp = pd.DataFrame(d)
    final_df = final_df.append(df_temp, ignore_index=True)
    
```
